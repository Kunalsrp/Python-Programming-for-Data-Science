```python
x = [-5,-2,-1,2,4,6,7,10]
y = [35,34,60,50,20,70,90,80]
plt.figure(figsize=(6,4))       #dimensions of figure in inches
plt.suptitle("Learning Visualisation",size=25)
plt.xlim(-20,20)
plt.ylim(10,100)
plt.plot(x,y)
```

```python
#plot graph of nos. from (-100 to 100) and its square . using range function
x = list(range(-100,100))
y = [i**2 for i in x]   #list comprehension
plt.plot(x,y)
```

```python
from math import exp 
x = list(range(0,30))
y = [exp(i) for i in x]   #list comprehension
plt.plot(x,y)
```

```python
from math import log 
x = list(range(1,101))
y = [log(i) for i in x]   #list comprehension
plt.plot(x,y)
```

```python
#plot multiple at a time : eg 12 graph at a time = 3x4
from math import *
plt.figure(figsize=(16,9)) 
x = list(range(-10,11))
y = [i**2 for i in x]
plt.subplot(3,4,1)        # 3 row 4 column and 1st position of grid
plt.plot(x,y)

y = [i**3 for i in x]
plt.subplot(3,4,8)        #  8th position of grid
plt.plot(x,y)

y = [sin(i) for i in x] 
plt.subplot(3,4,10)        #  10th position of grid
plt.plot(x,y)
```
